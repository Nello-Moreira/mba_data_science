```python
import pandas as pd
import numpy as np
import statsmodels.api as smf
```

```python
original_data = pd.read_excel('/app/src/datasets/companies.xlsx')
original_data
```

```python
original_data.describe()
```

```python
df = original_data.drop(columns=['ativo total.1'])
df = df.replace(to_replace='-', value=np.NaN).dropna()
df['dividendos'] = df['dividendos'].apply(lambda x: abs(x))
df
```

```python
df.describe()
```

```python
df = df.drop_duplicates(subset=['nome', 'ano'])
df
```

```python
df.describe()
```

```python
r = np.corrcoef(df['lucro liquido'], df['dividendos'])
pd.DataFrame(r, columns=['lucro liquido', 'dividendos'])
```

```python
x = smf.add_constant(df[['caixa', 'dividendos', 'patrimonio liquido']])
smf.regression.linear_model.OLS(df['valor mercado'], x).fit().summary()
```
