```python
import pandas as pd
import statsmodels.api as smf
```

```python
original_data = pd.read_excel('/app/src/datasets/products_report.xlsx')
original_data
```

```python
df = original_data
```

```python
df[['Quantidade',	'Preço de venda',	'Desconto',	'Custo']].corr()
```

```python
x = smf.add_constant(df[['Quantidade', 'Preço de venda', 'Desconto']])
smf.regression.linear_model.OLS(df['Custo'], x).fit().summary()
```
