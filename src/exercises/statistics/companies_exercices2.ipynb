```python
import pandas as pd
import numpy as np
import scipy as sp
import statsmodels.api as smf
import math
```

```python
original_data = pd.read_excel('/app/src/datasets/companies_2.xlsx')
original_data
```

```python
df = original_data.drop(columns=[original_data.columns[-1]])
df.columns = list(map(lambda x: x.replace('/ ', ''),map(lambda x: x.replace('\n', ' '), list(df.columns))))
df = df.replace(to_replace='-', value=0).replace(' ', 0)
df
```

```python
df.describe()
```

### Correlações

```python
df[['Mrg Ebitda',	'Margem Liquida',	'Valor Mercado',	'EV',	'ROA',	'ROE', 'PL',	'RECEITA',	'LUCRO BRUTO',	'EBIT',	'LUCRO LIQUID']].corr()
```

### ROA vs ROE

```python
sp.stats.ttest_ind(df['ROA'], df['ROE'])
```

```python
r = np.corrcoef(df['ROA'], df['ROE'])
pd.DataFrame(r, columns=['ROA', 'ROE'])
```

### mrg ebitda vs mrg liquida

```python
sp.stats.ttest_ind(df['Mrg Ebitda'], df['Margem Liquida'])
```

```python
r = np.corrcoef(df['Mrg Ebitda'], df['Margem Liquida'])
pd.DataFrame(r, columns=['Mrg Ebitda', 'Margem Liquida'])
```

### ebit vs lucro liquido

```python
sp.stats.ttest_ind(df['EBIT'], df['LUCRO LIQUID'])
```

```python
r = np.corrcoef(df['EBIT'], df['LUCRO LIQUID'])
pd.DataFrame(r, columns=['EBIT', 'LUCRO LIQUID'])
```

### Regressão margem ebitda VS ROE

```python
x = smf.add_constant(df[['ROE']])
smf.regression.linear_model.OLS(df['Mrg Ebitda'], df[['ROE']]).fit().summary()
```

### Regressão margem ebitda VS ROE + log ativo

```python
df['log_ativo'] = df['Ativo']
df.loc[df['log_ativo']>0, 'log_ativo'] = df['log_ativo'].loc[df['log_ativo']>0].map(lambda x: math.log(x))
df['log_ativo']
```

```python
x = smf.add_constant(df[['ROE', 'log_ativo']])
smf.regression.linear_model.OLS(df['Mrg Ebitda'], x).fit().summary()
```
